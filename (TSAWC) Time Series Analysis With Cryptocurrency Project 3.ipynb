```python
from statsmodels.tsa.stattools import adfuller

# Apply ADF test
result = adfuller(bitcoin_df['Close'])

# Print results
print("ADF Statistic:", result[0])
print("p-value:", result[1])
print("Critical Values:", result[4])

# Interpretation
if result[1] > 0.05:
    print("The series is Non-Stationary (Needs Differencing).")
else:
    print("The series is Stationary.")
```

```python
# First-order differencing
bitcoin_df['Close_diff'] = bitcoin_df['Close'].diff()

# Drop NA values created by differencing
bitcoin_df.dropna(inplace=True)

# Re-run ADF Test
result_diff = adfuller(bitcoin_df['Close_diff'])

# Print results
print("ADF Statistic (Differenced):", result_diff[0])
print("p-value:", result_diff[1])

# Interpretation
if result_diff[1] < 0.05:
    print("The differenced series is stationary (reject H0).")
else:
    print("The differenced series is still non-stationary (consider further differencing).")
```

# ARIMA

```python
from pmdarima import auto_arima

# Fit auto ARIMA model
stepwise_fit = auto_arima(bitcoin_df['Close_diff'], 
                          seasonal=False, 
                          trace=True, 
                          suppress_warnings=True)

# Print the best (p, d, q) values
print("Optimal ARIMA Order:", stepwise_fit.order)
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
model = ARIMA(bitcoin_df['Close_diff'], order=(2, 0, 1))
arima_result = model.fit()

# Forecast
forecast_steps = 30
forecast = arima_result.forecast(steps=forecast_steps)

# Plot results
plt.figure(figsize=(12,6))
plt.plot(bitcoin_df.index[-100:], bitcoin_df['Close_diff'].iloc[-100:], label='Actual', color='blue')
plt.plot(pd.date_range(bitcoin_df.index[-1], periods=forecast_steps, freq='D'), forecast, label='Forecast', color='red')
plt.legend()
plt.title("Bitcoin Price Forecast using ARIMA(2,0,1)")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Get actual values for comparison (last 30 points)
actual_values = bitcoin_df['Close_diff'].iloc[-30:]

# Compute error metrics
mae = mean_absolute_error(actual_values, forecast)
mse = mean_squared_error(actual_values, forecast)
rmse = mse ** 0.5

# Print results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
# Reverse differencing to get actual forecasted values
last_actual_close = bitcoin_df['Close'].iloc[-30]  # Last known actual closing price
forecast_actual = last_actual_close + forecast.cumsum()  # Reverse differencing

# Create DataFrame for visualization
forecast_df = pd.DataFrame({
    'Date': pd.date_range(start=bitcoin_df.index[-30], periods=30, freq='D'),
    'Forecast_Close': forecast_actual
})
```

# LSTM

```python
from sklearn.preprocessing import MinMaxScaler

# Scaling the data for better LSTM performance
scaler = MinMaxScaler(feature_range=(0,1))
bitcoin_scaled = scaler.fit_transform(bitcoin_df[['Close']])

# Creating sequences for LSTM model
def create_sequences(data, seq_length):
    sequences, labels = [], []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i + seq_length])
        labels.append(data[i + seq_length])
    return np.array(sequences), np.array(labels)

seq_length = 60  # Use past 60 days to predict the next day
X, y = create_sequences(bitcoin_scaled, seq_length)

# Splitting into training and testing sets
split = int(len(X) * 0.8)  
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Reshaping for LSTM input (samples, timesteps, features)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

print("LSTM Data Preparation Complete! ✅")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Building LSTM Model
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(seq_length, 1)),
    Dropout(0.2),
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),
    Dense(units=25),
    Dense(units=1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

print("LSTM Model Training Complete! ✅")
```

```python
# Predict on test data
predictions = model.predict(X_test)

# Reverse scaling to get actual values
predictions = scaler.inverse_transform(predictions.reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(y_test_actual, label="Actual Prices", color='blue')
plt.plot(predictions, label="Predicted Prices", color='red')
plt.xlabel("Time")
plt.ylabel("Bitcoin Price")
plt.legend()
plt.title("Bitcoin Price Prediction using LSTM")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculate error metrics
mae = mean_absolute_error(y_test_actual, predictions)
mse = mean_squared_error(y_test_actual, predictions)
rmse = np.sqrt(mse)

# Display results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

# Anomaly Detection in Cryptocurrency Prices

```python
from scipy.stats import zscore

# Compute Z-scores
bitcoin_df['z_score'] = zscore(bitcoin_df['Close'])

# Set threshold (e.g., |z_score| > 3 as anomaly)
threshold = 3
bitcoin_df['anomaly_zscore'] = (abs(bitcoin_df['z_score']) > threshold)

# Plot anomalies
plt.figure(figsize=(12, 6))
plt.plot(bitcoin_df.index, bitcoin_df['Close'], label="Bitcoin Close Price", color='blue')
plt.scatter(bitcoin_df.index[bitcoin_df['anomaly_zscore']], 
            bitcoin_df['Close'][bitcoin_df['anomaly_zscore']], 
            color='red', label="Anomalies", marker='o')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Bitcoin Price Anomalies (Z-Score)")
plt.legend()
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = bitcoin_df['Close'].quantile(0.25)
Q3 = bitcoin_df['Close'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify anomalies
bitcoin_df['anomaly_iqr'] = (bitcoin_df['Close'] < lower_bound) | (bitcoin_df['Close'] > upper_bound)

# Plot anomalies
plt.figure(figsize=(12, 6))
plt.plot(bitcoin_df.index, bitcoin_df['Close'], label="Bitcoin Close Price", color='blue')
plt.scatter(bitcoin_df.index[bitcoin_df['anomaly_iqr']], 
            bitcoin_df['Close'][bitcoin_df['anomaly_iqr']], 
            color='red', label="Anomalies", marker='o')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Bitcoin Price Anomalies (IQR Method)")
plt.legend()
plt.show()
```
